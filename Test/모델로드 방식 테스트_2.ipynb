```python
from transformers import AutoModelForCausalLM, AutoTokenizer

model_name = "taeseo06/llama2-PersonalizedFinetuning"
model = AutoModelForCausalLM.from_pretrained(model_name).float()
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
input_text = "나는 오늘 텍스트 생성 모델을 사용하는 방법을 배우고 있습니다."


input_text = input_text[:100]  # 입력 텍스트를 100자로 제한

outputs = model.generate(
    tokenizer(input_text, return_tensors="pt").input_ids,
    max_length=200,
    num_beams=4,
    no_repeat_ngram_size=3,
    temperature=1.0,
    top_k=50,
    eos_token_id=tokenizer.eos_token_id,
)

predicted_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

print(f"생성된 텍스트: {predicted_text}")

```

```python
def get_response(prompt: str) -> None:
    """
    Generate a response from the Llama model.

    Parameters:
        prompt (str): The user's input/question for the model.

    Returns:
        None: Prints the model's response.
    """
    sequences = pipe(
        prompt,
        do_sample=True,
        top_k=10,
        num_return_sequences=1,
        eos_token_id=tokenizer.eos_token_id,
        max_length=256,
    )
    print("Chatbot:", sequences[0]['generated_text'])
```
