```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("text-generation", model="beomi/KoAlpaca-KoRWKV-6B")
```

```python
def gen(x):
    a = pipe(
        f'### 질문: {x}\n\n### 답변:',
        max_new_tokens=1024,
        return_full_text=False,
        do_sample=True,
        top_p=0.9,
        temperature=0.7,
        early_stopping=True,
        repetition_penalty=1.2,
    )
    print(a[0]['generated_text'])
```

```python
gen("오늘 기분은 좀 어때?")
```
